For gender column, we should fill the value with:
mode - most frequent value - male
least frequent value - female
respective ratios

Here I am using Mode(max count) to fill inplace of missing values

```python
df['gender'].fillna(df['gender'].mode()[0],inplace=True) 
df['gender']
```

```python
df
```

```python
df.isnull().sum()
```

-I have replaced missing in gender column now remaining is two columns  

```python
 df['maximum_days_inactive'].median()
```

```python
df['maximum_days_inactive'].mean()
```

here I can replace with either mean or median according to your assumption or thinking as median gives me perfect whole number i go for median

```python
df.fillna({'maximum_days_inactive': df['maximum_days_inactive'].median()}, inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

here I have replaced Maximum_days_inactive column  with median of that column so remaining one column has missing value 

```python
df['churn'].value_counts(dropna=False)
```

1.0->leave
0.0->stay

this column says where the customer(user) contiues(stay) or discontinues(leave) from this OTT platform

Here i am replacing the missing values by frequency(most repeated occurance) using mode() according to 
my analysis  based on CUSTOMER CALL column and some other column



```python
df['churn'].fillna(df['churn'].mode()[0],inplace=True) 
df['churn']
```
